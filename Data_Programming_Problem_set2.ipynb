# Question 1

### Code Solution :

```python
a = 0
def b():
    global a
    a = c(a)
def c(a):
    return a + 2

b()
b()
b()
a
```

### Explanation :

Answer : The final value of <b><i>a</i></b> is 6.

- In the first line of code, <b><i>a</i></b> is initiated with a value of 0.


- In <i>function definition of b</i>, the <b>global</b> keyword is added with <b><i>a</i></b> because we cannot modify a <b>global variable</b> inside function, later <b><i>a</i></b> is assigned value of <b><i>c(a)</i></b>, which is then calling the <i>function c</i>


- In <i>function definition of c</i>, the value of <b><i>a</i></b> is then incremented by 2, since we added <b>global</b> keyword with <b><i>a</i></b> in the previous function, this will change the value of <b><i>a</i></b> globally (meaning outside the function too)

# Question 3

### Code Solution :

```python
#class named "Marsupial" is created.

class Marsupial: 
    def __init__(self): #__init__ method initializes the attributes of each instance of class
        self.pouch_list = [] #pouch_list is initialized empty, so that later elements can be added into it.
        
    def put_in_pouch(self,element): #function to put element entered by user in pouch_list
        self.element = element # 
        self.pouch_list.append(self.element) # appending element to list
        
    def pouch_contents(self): #function to print all the elements in list
        return self.pouch_list        
```

```python
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())
```

```python
#Creating class named "Kangaroo" that inherits from "Marsupial"

class Kangaroo(Marsupial): 
    def __init__(self,x,y): #creating new __init__ method that will extend the parent's __init__ method 
        super().__init__() #this line of code inherit all the attributes and function from super class i.e. parent class
        self.x=x
        self.y=y
        
    def jump(self,dx,dy): #creating jump function, that takes 2 inputs from user
        self.dx=dx
        self.dy=dy
        self.x=self.x+self.dx #adding dx value to initial x position of instance
        self.y=self.y+self.dy #adding dy value to initial y position of instance
        
    def __str__(self): # overwiting the __str__ function to display the result in string format
        return "I am a Kangaroo located at coordinates ({},{})".format(self.x,self.y)
```

```python
k = Kangaroo(0,0)
print(k)
```

```python
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())
```

```python
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

```python
def collatz(x): #creating function named "collatz" that takes input entered by user
    
    if x==1:  # base case, if x becomes 1, print (1) and end the recursive loop
        print(int(x))
        
    else:
        print(int(x)) #if x !=1, print value of x . for ex: at begininning value of x is 10
        
        if x%2==0: #then checking the conditions, if the value of x is even
            collatz(x/2) #call function again by passing x/2 to function
            
        elif x%2!=0: #if value of x is odd
            collatz(3*x+1) #call dunction again by passing 3*x+1 to function
            
collatz(10)
        
```

# Question 5

### Code Solution :

```python
#Craeting the rrecursive function named "binary" that takes user input

def binary(x): #it includes 2 base cases
    
    if x==0: #if value of x is 0 print 0
        return 0
        
    elif x==1: #if value of x becomes 1 print 1
        return 1;
        
    else:
        return str(binary(x//2)) + str(x%2) #call the function again with x//2. for example: x=3, binary(1) is called
        

print("Binary equivalent of {} = {}".format(0,binary(0)))
print("Binary equivalent of {} = {}".format(1,binary(1)))
print("Binary equivalent of {} = {}".format(3,binary(3)))
print("Binary equivalent of {} = {}".format(9,binary(9)))
print("Binary equivalent of {} = {}".format(23,binary(23)))
```

### Code Solution :

#### Creating the database and the table given first

```python
import sqlite3 

con = sqlite3.connect('temp.db') # making connection and creating new database named "temp"
cur = con.cursor() #initiating the cursor, that will help us to execute SQL statements

#executing the first SQL query to create table called "Temperature" with column names as city,country,season,temp,rainfall
cur.execute("CREATE TABLE Temperature (City, Country, Season, Temp, Rainfall)") 
```

```python
#running a for loop to add the values in table "Temperature"

for x in [["Mumbai","India","Winter",24.8,5.9],
          ["Mumbai","India","Spring",28.4,16.2],
          ["Mumbai","India","Summer",27.9,1549.4],
          ["Mumbai","India","Fall",27.6,346.0],
          ["London","United Kingdom","Winter",4.2,207.7],
          ["London","United Kingdom","Spring",8.3,169.6],
          ["London","United Kingdom","Summer",15.7,157.0],
          ["London","United Kingdom","Fall",10.4,218.5],
          ["Cairo","Egypt","Winter",13.6,16.5],
          ["Cairo","Egypt","Spring",20.7,6.5],
          ["Cairo","Egypt","Summer",27.7,0.1],
          ["Cairo","Egypt","Fall",22.2,4.5]]:
    cur.execute("INSERT INTO Temperature VALUES (?,?,?,?,?)",x)
```

```python
con.commit()
con.close()

#below is the final table stored in database
```

### a) All the temperature data.

```python
con = sqlite3.connect('temp.db')
cur = con.cursor()
cur.execute('SELECT Temp FROM Temperature')
print("a) ALL the temperature data:\n")
for i in cur.execute('SELECT Temp FROM Temperature'):
    print(i[0])
```

*******************************************************************************************************************************

### b) All the cities, but without repetition.

```python
cur.execute('SELECT Distinct City FROM Temperature')
print("b) All the cities, but without repitition:\n")
for i in cur.execute('SELECT Distinct City FROM Temperature'):
    print(i[0])
```

*******************************************************************************************************************************

### c) All the records for India

```python
cur.execute('SELECT * FROM Temperature WHERE Country = "India"')
print("c) All the records for India:\n")
for i in cur.execute('SELECT * FROM Temperature WHERE Country = "India"'):
    print(i)
```

*******************************************************************************************************************************

### d) All the Fall records

```python
cur.execute('SELECT * FROM Temperature WHERE Season = "Fall"')
print("d) All the Fall records:\n")
for i in cur.execute('SELECT * FROM Temperature WHERE Season = "Fall"'):
    print(i)
```

*******************************************************************************************************************************

#### e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.

```python
cur.execute('SELECT City,Country,Season FROM Temperature WHERE Rainfall Between 200 and 400')
print("e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters:\n")
for i in cur.execute('SELECT City,Country,Season FROM Temperature WHERE Rainfall Between 200 and 400'):
    print(i)
```

*******************************************************************************************************************************

#### f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.

```python
cur.execute('SELECT City,Country FROM Temperature WHERE Season = "Fall" and Temp > 20 ORDER BY Temp')
print("f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.:\n")
for i in cur.execute('SELECT City,Country FROM Temperature WHERE Season = "Fall" and Temp > 20 ORDER BY Temp'):
    print(i)
```

*******************************************************************************************************************************

### g) The total annual rainfall for Cairo.

```python
cur.execute('SELECT SUM(Rainfall) from Temperature Where City= "Cairo"')
print("g) The total annual rainfall for Cairo.:\n")
for i in cur.execute('SELECT SUM(Rainfall) from Temperature Where City= "Cairo"'):
    print(i[0])
```

### h) The total rainfall for each season.

```python
cur.execute('SELECT SUM(Rainfall), Season from Temperature Group BY Season')
print("h) The total rainfall for each season:\n")
for i in cur.execute('SELECT SUM(Rainfall), Season from Temperature Group BY Season'):
    print(i)
```

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over','the', 'lazy', 'dog'] #initialy the words list looks like this

#Using List Comprehensions to perform all the task

print("a) {}\n".format([i.upper() for i in words]))
print("b) {}\n".format([i.lower() for i in words]))
print("c) {}\n".format([len(i) for i in words]))
print("d) {}\n".format([[i.upper(),i.lower(),len(i)] for i in words]))
print("e) {}\n".format([i for i in words if len(i)>=4]))
```
