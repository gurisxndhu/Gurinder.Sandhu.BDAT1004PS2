```python
int a,b,c; # taking three variables to store the lengths of three sides of triangle

Console.WriteLine("\n\t\t\tEnter the length of 3 sides of triangle below: \n"); # taking user's input

Console.Write("Enter length of first Side:");
a = Convert.ToInt32(Console.ReadLine()); 

Console.Write("Enter length of second Side:");
b = Convert.ToInt32(Console.ReadLine());

Console.Write("Enter length of third Side:");
c = Convert.ToInt32(Console.ReadLine());

Console.WriteLine($"\nBy Heron's Formula, Area of triangle with sides {a},{b},{c} is {triangleArea(a,b,c)}");


double triangleArea(int a, int b, int c)
{
    int s = (a + b + c) / 2;
    double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c)); # just applying the formula mentioned and returning the area
    return area;
}
```

# Question 4 C#

### Code Solution :

```python
int size; # creating size variable to store size of array

Console.Write("Enter the number of elements to be stored in array:"); 
size = Convert.ToInt32(Console.ReadLine()); # storing user's input of size by typcasting the input into <int>

Console.WriteLine($"\nInput {size} elements in the array:\n");

List<int> array = new List<int>(); #initializing list array[], that stores all the values entered by user

List<int> odd = new List<int>(); #initializing list odd[], stores odd values present in array[]

List<int> even = new List<int>(); #initializing list even[], stores even values present in array[]

for (int i=0; i<size; i++) # interating and storing the values in array[] list entered by user
{
    Console.Write($"Element {i}: ");
    array.Add(Convert.ToInt32(Console.ReadLine()));
}

for (int i=0 ; i< array.Count; i++) # iterating through the complete array[]
{
    if (array[i] %2==0) # if value (mod) 2 ==0 means diving value by 2 gives remainder as 0, it is even
    {
        even.Add(array[i]); # storing that value in even[] list
    }
    else
    {
        odd.Add(array[i]); # else storing other values in odd[] list
    }
}

Console.WriteLine("The Even element are: ");

for (int i=0; i<even.Count;i++) # iterating over even[] list and printing even values
{
    Console.Write($"{even[i]} ");
}

Console.WriteLine("\nThe Odd elements are: "); # iterating over odd[] list and printing odd values

for (int i=0; i< odd.Count;i++)
{
    Console.Write($"{odd[i]} ");
}
```

# Question 5 C#

### Code Solution :

```python
Console.WriteLine($"inside(1, 1, 0, 0, 2, 3) : {inside(1, 1, 0, 0, 2, 3)}");
Console.WriteLine($"inside(-1, -1, 0, 0, 2, 3) : {inside(-1, -1, 0, 0, 2, 3)}");
Console.WriteLine($"inside(1, 1, 0.3, 0.5, 1.1, 0.7) : {inside(1, 1, 0.3, 0.5, 1.1, 0.7)}");
Console.WriteLine($"inside(1, 1, 0.5, 0.2, 1.1, 2) : {inside(1, 1, 0.5, 0.2, 1.1, 2)}");

"""
We need to check if (x,y) is inside rectagle with lower left (x1,y1) and upper right (x2,y2),
it is true only if both the conditions below are true

1. if x is between x1 and x2 : x1 <= x <= x2
2. if y is between y1 and y2 : y1 <= y <= y2

"""

bool inside(double x, double y, double x1, double y1, double x2, double y2)
{
    bool result;
    if ((x >= x1 & x <= x2) & (y >= y1 & y <= y2)) 
    {
        result = true;
    }
    else
    {
        result = false;
    }

    return result;

}
```

# Question 6 Python

### Code Solution :

```python
def pig(userInput): # fuction that takes user's string input as argument
    
    vowels = ['a','e','i','o','u'] # storing all the vowels in a list
    
    if userInput.lower()[0] in vowels: # applying lowercase on user input and checking if first char is vowel
        userInput +="way" # if yes then appending "way" to the end of input
    else:
        userInput+=userInput[0]+"ay" # else adding char at 0 index and "ay" -> happy = happyhay
        userInput = userInput[1:] # now slicing the result to remove the first char -> happyhay = appyhay

    return userInput.lower()

print(pig("happy"))    
print(pig("Enter")) 
print(pig("pencil")) 
print(pig("other")) 
```

# Question 7 Python

### Code Soution :

```python
def bldcount(file):
    bloodGroups = ["A", "B", "AB", "O", "OO"] #Storing all the bloodtypes in list 
    
    bloodCount = {} #Creating count Dictionary to store bloodtype as key and its count as value -> {"A" :15}
    
    f = open(file,"r") # reading the file that is passed as argument in function
    
    for blood in f.read().split(): # splitting the text file by space and storing each record in list and iterating over it
        
        if blood in bloodCount.keys(): # if the record is already present in dictionary
            bloodCount[blood]+=1 # we will increment the count by 1 
        else:
            bloodCount[blood] = 1 # else we will create new key with that record and initiate its count as 1
        
    
    """ The folowing block of code check if all the bloodGroups present in list have been added to dictionary,
        if any bloodGroup is missing in Dictionary means it never occurred in text file, so we create new key
        for that record and store its count as 0 """
    for x in bloodGroups:
        if x not in bloodCount.keys():
            bloodCount[x]=0
                   
    print("There {} {} {} of blood type A.".format("is" if bloodCount["A"] == 1 else "are", "no" if bloodCount["A"]==0 else bloodCount["A"] , "patient" if bloodCount["A"] == 1 else "patients"))
    print("There {} {} {} of blood type B.".format("is" if bloodCount["B"] == 1 else "are", "no" if bloodCount["B"]==0 else bloodCount["B"] , "patient" if bloodCount["B"] == 1 else "patients"))
    print("There {} {} {} of blood type AB.".format("is" if bloodCount["AB"] == 1 else "are", "no" if bloodCount["AB"]==0 else bloodCount["AB"] , "patient" if bloodCount["AB"] == 1 else "patients"))
    print("There {} {} {} of blood type O.".format("is" if bloodCount["O"] == 1 else "are", "no" if bloodCount["O"]==0 else bloodCount["O"] , "patient" if bloodCount["O"] == 1 else "patients"))
    print("There {} {} {} of blood type OO.".format("is" if bloodCount["OO"] == 1 else "are", "no" if bloodCount["OO"]==0 else bloodCount["00"] , "patient" if bloodCount["OO"] == 1 else "patients"))
    
    
bldcount("bloodtype1.txt") # Calling the function by passing text file containing records of patients
```

# Question 8 Python

### Code Solution :

```python
def curconv(currency,amount): # fuction that takes two arguments - currency and amount
    
    f = open("currencies.txt","r") #storing the currency.txt file in f variable
    
    
    for i in f.readlines(): # reading the text file line by line
        
    """ As we know each line in text file contains abbreviation for currency, its conversion rate, and complete
        currency name, we will split them by space and store each value in list "rates[]" where,
        rates[0] = abbreviation for currency
        rates[1] = its conversion rate
        rates[2] = complete currency name
        """
        rates= i.split() 
        
        if rates[0] == currency: # if value at 0 index of rates matches the currency passed in func
            result = float(rates[1])*amount # we will extract its conversion rate which is stored at index 1 and multiply it with the amount entered by user
    return result
        
print(curconv('EUR',100))    
print(curconv('JPY',100))  
print(curconv('THB',100))
```

# Question 10 Python


### Code Solution :

```python
def frequencies(stringInput): # function taking userInput as string
    letters = "abcdefghijklmnopqrstuvwxyz" #all the lowercase alphabets are stored in string "letters"
    count = {} #initializing count dictionary to store the count of each char occuring in stringInput

    for i in letters:
        count[i]=0 # Storing starting count as 0 for all chars in count dictionary
        
    for c in stringInput: # iterating through user's input string
        if c in count: # if char is found in dictionary
            count[c]+=1 # we increment the count by 1
        else:
            continue # else we will go to next char
    
    return list(count.values()) # count.values is of type <dict_values> so we typecast it into <list>
        
h = input("Enter a string:")
print()
print("The quick red fox got bored and went home. :")
print(frequencies("The quick red fox got bored and went home."))
print()
print("apple :")
print(frequencies("apple"))
print()
print("{} :".format(h))
print(frequencies(h))
```
